# Python Realization of FFT

```python
import numpy as np
from scipy.fftpack import fft, ifft
import matplotlib.pyplot as plt
from matplotlib.pylab import mpl


#select 1,400 points
x=np.linspace(0, 1, 1400)
print("The array:")
print(x)

#set the signal (waveforms); different frequencies: 200Hz, 400 and 600Hz.
y=7*np.sin(2*np.pi*200*x) + 5*np.sin(2*np.pi*400*x)+3*np.sin(2*np.pi*600*x)

#Draw the figure:
plt.figure()
plt.plot(x, y)
plt.title("Original waveforms")

plt.figure()
plt.plot(x[0:50], y[0:50])
plt.title("First 50 points")
plt.show()

#Simple application of FFT:
fft_y = fft(y)
print(len(fft_y))
print(fft_y[0:10])

#Original frequency figure of FFT:
N=1400
x=np.arange(N)

abs_y=np.abs(fft_y)
angle_y=np.angle(fft_y)

plt.figure()
plt.plot(x, abs_y)
plt.title("Distribution of absolute values of FFT")

plt.figure()
plt.plot(x, angle_y)
plt.title("Distribution of phases of FFT")

#Normalization:
normalization_y = abs_y/N
plt.figure()
plt.plot(x, normalization_y, 'g')#g: green????????????????
plt.title("Normalized", fontsize=16, color='green')#color here is related to title
plt.show()

#keep half of the figure:
half_x = x[range(int(N/2))]
normalization_half_y = normalization_y[range(int(N/2))]
plt.figure()
plt.plot(half_x, normalization_half_y, 'r')
plt.title("Half", fontsize=16, color='red')
plt.show()
```
